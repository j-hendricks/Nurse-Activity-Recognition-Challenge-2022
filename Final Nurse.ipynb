```python

activity_type_id = range(1,29)
activity_type = ["vital","Meal/Med","Oral Care","Excretion","Bathing/Wiping","Treatment","Morning Gathering/Excercise","Rehab/Recreation","Morning Care", "Daytime User Response", "Night Care", "Nighttime User Response","Family/Guest Response","Outing Response","Linen Exchange", "Cleaning", "Handwriting Recording","Delegating/Meeting","Get Up Assistance", "Change Dressing Assistance", "Washing Assistance","Doctor Visit Correspindence","Preparation and Checking of Goods","Organization of Medications","Family Doctor Contact","Break","Emergency Response","Special Remarks/Notes"]
```

```python
care['start'] = pd.to_datetime(care['start'], format='%Y-%m-%d %H:%M:%S%z')
care['finish'] = pd.to_datetime(care['finish'], format='%Y-%m-%d %H:%M:%S%z')
care['year-month-date'] = care['start'].dt.date.astype(str)
care['hour'] = care['start'].dt.hour
care['minute'] = care['start'].dt.minute
care['year-month-date-hour'] = care['year-month-date'].astype(str) + '-' +care['hour'].astype(str).str.zfill(2)
care = care.sort_values('start')
```

```python
care["Duration"] = (care["finish"] - care["start"]).dt.total_seconds()
care['isMorning'] = (care.hour > 6) & (care.hour < 10)
care['isDay'] = (care.hour >= 10) & (care.hour < 16)
care['isEvening'] = (care.hour >= 16) & (care.hour < 20)
care['isNight'] = (care.hour >= 20) | (care.hour < 6)
care
```

```python
def makeFeatures(df):
    feat = df.groupby(['activity_type_id', 'year-month-date-hour'], as_index=False).count()[['activity_type_id', 'year-month-date-hour','id']].rename(columns={'id':'count'})
    X = care.copy()
    acts = list(df['activity_type_id'].value_counts().index)
    acts.sort()
    print(acts)
    for act in acts:
        X = pd.merge(X, feat[feat['activity_type_id'] == act][['year-month-date-hour', 'count']], on='year-month-date-hour', how="left").rename(columns={'count':str(act)})
    return X

tmp_df = care.copy()
features_cr_df = makeFeatures(tmp_df)
features_cr_df = features_cr_df.fillna(0).sort_values('year-month-date-hour')
care = features_cr_df.reset_index()
```
